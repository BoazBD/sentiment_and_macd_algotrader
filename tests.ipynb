```python
import concurrent.futures

def task1():
    print("Thread 1: Starting task 1...")
    time.sleep(2)  # Simulating some work for Thread 1
    print("Thread 1: Task 1 completed!")
    return "Result from task 1"

def task2():
    print("Thread 2: Starting task 2...")
    time.sleep(6)  # Simulating some work for Thread 2
    print("Thread 2: Task 2 completed!")
    return "Result from task 2"

def main():
    # Create a ThreadPoolExecutor with two threads
    with concurrent.futures.ThreadPoolExecutor(max_workers=2) as executor:
        # Submit the tasks to the executor
        future1 = executor.submit(task1)
        future2 = executor.submit(task2)

        # Wait for Thread 2 to finish and get the result of its task
        result1 = future1.result()
        result2 = future2.result()

    print("Main thread: Both threads completed!")
    print("Result from task 1:", result1)
    print("Result from task 2:", result2)

if __name__ == '__main__':
    main()
```

```python
import requests
from bs4 import BeautifulSoup

def get_sp500_data():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

    # Send a GET request to the Wikipedia page
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the table containing the company data
    table = soup.find("table", {"class": "wikitable sortable"})

    tickers = []
    company_names = []

    # Iterate over each row in the table, skipping the header row
    for row in table.find_all("tr")[1:]:
        # Get the ticker and company name from the first and second columns
        columns = row.find_all("td")
        ticker = columns[0].text.strip()
        name = columns[1].text.strip()

        tickers.append(ticker)
        company_names.append(name)

    return tickers, company_names

# Get the tickers and company names of S&P 500 companies
sp500_tickers, sp500_names = get_sp500_data()

# Display the tickers and company names
for ticker, name in zip(sp500_tickers, sp500_names):
    print(ticker, "-", name)

#save it in a csv file\
import pandas as pd
df = pd.DataFrame({'Ticker':sp500_tickers, 'Company Name':sp500_names})
df.to_csv('CompanyNames2.csv', index=False)

```

```python
def find_stock_interest(stock_list, timeframe):
    pytrends = TrendReq(hl='en-US', tz=360)

    # Store the results for each stock
    results = []

    # API can only handle 5 keywords at a time, so split stocks into chunks of 5
    chunk_size = 5
    stock_chunks = [stock_list[i:i+chunk_size] for i in range(0, len(stock_list), chunk_size)]

    # Iterate over stock chunks and retrieve the interest over time
    # Calculate interest score every 8 minutes for 24 hours, and make an average for each stock 
    for chunk in stock_chunks:
        pytrends.build_payload(kw_list=chunk, timeframe=timeframe)
        interest_over_time_df = pytrends.interest_over_time()
        results.append(interest_over_time_df)
    
    
    # Concatenate the results and compute the maximum average of interest across all stocks
    combined_results = pd.concat(results, axis=1)
    return combined_results.mean().sort_values(ascending=False)
```
